```python
# Spam detection, using data from https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection

# Import data
import pandas as pd

dataset = pd.read_table('dataset/SMSSpamCollection',
                        sep='\t',
                        header=None,
                        names=['label', 'message'])

dataset.head()
```

```python
# Preprocessing data

# convert labels values into 0 or 1 (ham = 0, spam = 1)

dataset['label'] = dataset.label.map({'ham':0, 'spam':1})
print(dataset.shape)
dataset.head() # returns (rows, columns)
```

```python
# splitting data
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataset['message'], 
                                                    dataset['label'], 
                                                    random_state=1)

print('Number of rows in the total set: {}'.format(dataset.shape[0]))
print('Number of rows in the training set: {}'.format(X_train.shape[0]))
print('Number of rows in the test set: {}'.format(X_test.shape[0]))
```

```python
# Creating a Bag of word
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()


# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)

# Transform testing data and return the matrix. Note we are not fitting the testing data into the CountVectorizer()
testing_data = count_vector.transform(X_test)
```

```python
# convert couunt_vectorizer into a matrix with frequency
training_data = pd.DataFrame(training_data.toarray(), 
                                columns = count_vector.get_feature_names())
testing_data = pd.DataFrame(testing_data.toarray(), 
                                columns = count_vector.get_feature_names())
```

```python
# Creating classifier
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(training_data, y_train)
```

```python
# Predict our solution

predictions = clf.predict(testing_data)
```

```python
# Evaluating our model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
print('Precision score: ', format(precision_score(y_test, predictions)))
print('Recall score: ', format(recall_score(y_test, predictions)))
print('F1 score: ', format(f1_score(y_test, predictions)))
```
